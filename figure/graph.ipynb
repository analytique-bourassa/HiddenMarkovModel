```python
# libraries
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
# Build a dataframe with your connections
df = pd.DataFrame({ 'from':['State A', 'State B', 'State C', 'State A', 'State B', 'State C' ], 'to':['Emission-A', 'Emission-B', 'Emission-C','State B','State C','State A']})
df
 
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
fig = plt.figure(figsize=(12,10))
# Graph with Custom nodes:
nx.draw(G, with_labels=True, node_size=1200, node_color="skyblue", node_shape="s", alpha=0.5, linewidths=40)

plt.show()
```
